# Exercise 4.10

(Guess the Number) Write a script that plays “guess the number.” Choose the  number to be guessed by selecting a random integer in the range 1 to 1000. Do not reveal  this number to the user. Display the prompt "Guess my number between 1 and 1000 with  the fewest guesses:". The player inputs a first guess. If the guess is incorrect, display  "Too high. Try again." or "Too low. Try again." as appropriate to help the player “zero  in” on the correct answer, then prompt the user for the next guess. When the user enters  the correct answer, display "Congratulations. You guessed the number!", and allow the  user to choose whether to play again. 



```python
import random

# get a number between 1  to 1000.
rando= random.randint(1,1000)

while True:
    guess= input('Guess my number between 1 and 1000 with the fewest guesses:')
    if guess.isalpha():
        print('Please input a number.')
    else:
        if guess.isnumeric():
            guess=int(guess)
            if guess==rando:
                print("Congratulations. You guessed the number!")
                again=input('Play again? y or n:' )
                if ((again=='n')|(again=='N')|(again=="y")|(again=='Y')):
                    if((again=='n')| (again=='N')):
                        print('Great Game!')
                        break
                    else:
                        print( 'Lets play again!')
                        rando= random.randint(1,1000)
            elif guess<rando:
                print("Too low.Try again")
            elif guess>rando:
                print("Too High. Try again")
   
    
```
