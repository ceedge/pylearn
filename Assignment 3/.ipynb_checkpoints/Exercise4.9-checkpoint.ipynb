# Exercise 4.9

(Temperature Conversion) Implement a fahrenheit function that returns the  Fahrenheit equivalent of a Celsius temperature. Use the following formula:  
F = (9/ 5) * C + 32  

Use this function to print a chart showing the Fahrenheit equivalents of all Celsius temperatures in the range 0â€“100

 

```python
def fahrenheit(c):
    f=(9/5)*c+32
    return f

print(f'{"Celsius"}\t {"Fahrenheit"}')
print('-----','\t', '------')
for i in range (0,101):
    print(f'{i}\t {round(fahrenheit(i),1)}')
```
